13. Buscar el deporte en el que compitieron más deportistas, luego responder:
    1. Cuál fue el país que más deportistas contribuyo a este deporte?
    2. Cuál es el que obtuvo más medallas?

```python
deporte_max_ateltas = data[["Athlete","Sport"]].\
    drop_duplicates().\
    groupby(["Sport"]).\
    count().\
    reset_index().\
    sort_values("Athlete", ascending = False).\
    iloc[0,0]

print(f"El deporte con más atletas es {deporte_max_ateltas}")

# Respuesta A
top_pais_atletas = data.\
    loc[data["Sport"] == deporte_max_ateltas,["Country", "Athlete"]].\
    drop_duplicates().\
    groupby("Country").\
    count().\
    reset_index().\
    sort_values("Athlete", ascending = False).\
    iloc[0,:]

print(f"El país con más atletas en el deporte {deporte_max_ateltas} es {top_pais_atletas.iloc[0]} con {top_pais_atletas.iloc[1]} atletas")

# Respuesta B
top_pais_medallas = data.\
    loc[data["Sport"] == deporte_max_ateltas,["Country", "Total Medals"]].\
    groupby("Country").\
    sum().\
    reset_index().\
    sort_values("Total Medals", ascending = False).\
    iloc[0,:]

print(f"El país con más medallas en el deporte {deporte_max_ateltas} es {top_pais_medallas.iloc[0]} con {top_pais_medallas.iloc[1]} medallas")
```

17. Cuales son los nombres más largos de los ateltas en cuanto a cantidad de palabras? Este puede ser un tanto desafiante, las funciones split y join pueden servirte

```python
# opción con bucle for
max = 0
nombre = {}
for i in list(medals.Athlete.dropna().str.split(" ")): # divido las palabras, previamente elimino na para que no rompa
    if len(i) > max:
        max = len(i)
        nombre.clear()
        nombre[' '.join(s for s in i)] = len(i)
    elif len(i) == max:
        nombre[' '.join(s for s in i)] = len(i)
        
nombre
```

18. Calcular el promedio de medallas de bronce, plata y oro por deporte y país de los países que hayan recibido más medallas de oro que de plata.

```python
# opcion 1
data_agrupada = data.\
    groupby(["Sport", "Country"])[["Bronze Medals", "Silver Medals", "Gold Medals"]].\
    agg(["sum", "mean"]).\
    reset_index()

print(data_agrupada.columns) # vemos que las columnas ahora son indices múltiples, podemos llamar a las columnas nombrando toda la tupla 

filtro = data_agrupada[("Gold Medals", "sum")] > data_agrupada[("Silver Medals", "sum")]
data_agrupada.loc[filtro,:]
```

```python
# opcion 2
data_agrupada = data.\
    groupby(["Sport", "Country"])[["Silver Medals", "Gold Medals"]].\
    sum().\
    reset_index()

print(data_agrupada.columns) # vemos que las columnas ahora son indices normales

filtro = data_agrupada["Gold Medals"] > data_agrupada["Silver Medals"]
deportes_y_paises = data_agrupada.loc[filtro,["Sport", "Country"]]

data.\
    merge(deportes_y_paises, how = "inner").\
    groupby(["Sport", "Country"])[["Bronze Medals", "Silver Medals", "Gold Medals"]].\
    mean().\
    reset_index()
```

21. Cual fue el deportista más joven en obtener 10 medallas

```python
# agrupo por año y atleta
data_agrupada = data.\
    groupby(["Athlete", "Age"])["Total Medals"].\
    sum().\
    reset_index().\
    sort_values(["Athlete", "Age"], ascending = True)

# agrego una columna con el acumulado de medallas
data_agrupada["Medallas Acumuladas"] = data_agrupada.\
    groupby("Athlete").\
    agg({"Total Medals":"cumsum"})

# filtro todos los valores mayores o iguales a 10 medallas acumuladas y ordeno de menor a mayor por año
data_filtrada = data_agrupada.\
    loc[data_agrupada["Medallas Acumuladas"] >= 10, :].\
    sort_values("Age")
# se puede ver que los primeros dos consiguieron más de 10 medallas a los 23 años, podemos, agregando un filtro a lo anterior, obtener esta tabla
data_filtrada.\
    loc[data_filtrada["Age"] == data_filtrada["Age"].min(), :].\
    sort_values("Age")
```
